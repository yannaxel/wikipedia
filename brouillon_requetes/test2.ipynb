```python
import requests

S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

id = 13828397
PARAMS = {
    "action": "query",
    "format": "json",
    "prop": "categories",
    "pageids": id
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

PAGES = DATA["query"]["pages"]

for k, v in PAGES.items():
    for cat in v['categories']:
        print(cat["title"])
```

```python
S = requests.Session()

URL = "https://en.wikipedia.org/w/api.php"

PARAMS = {
    "action": "query",
    "format": "json",
    "list": "pagepropnames",
    "acfrom": "List of people and organisations named in the Paradise Papers"
}

R = S.get(url=URL, params=PARAMS)
DATA = R.json()

CATEGORIES = DATA["query"]["pagepropnames"]

for cat in CATEGORIES:
    print(cat["propname"])
```
